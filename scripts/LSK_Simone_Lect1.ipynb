## 리스트에서 가장 긴 단어 찾기

```python
words = ['hello', 'computer', 'bookshelves', 'chair', 'bicycle']

longest = len(words[0])
longest_word = words[0]

for word in words[1:]:
    if len(word) > longest:
        longest_word = word
        longest = len(word)
print("Longest word:", longest_word)
print("Length of the word:", longest)
```

## List Comprehension

```python
a = [ x**2 for x in range(10) ]
print(a)
```

```python
for x in range(10):
    print(x**2, end=" ")
```

```python
b = [ 2**b for b in range(21) ]
print(b)
```

```python
for b in range(21):
    print(2**b, end=' ')
```

```python
f = open('data/alice.txt')
text = f.read()
f.close()
words = text.split()

longest = len(words[0])
longest_word = words[0]

for word in words[1:]:
    if len(word) > longest:
        longest_word = word
        longest = len(word)
        
print("Longest word:", longest_word)
print("Length of the word:", longest)
```

## Word Cloud

```python
text = "Sungshin Sungshin Sungshin Sungshin University University"

import wordcloud
wc_obj = wordcloud.WordCloud()
wc = wc_obj.generate(text)

wc.words_
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.imshow(wc)
plt.axis("off")
plt.show()
```

```python
plt.savefig("wc_sungshin.png")
```

```python
import wordcloud
text = open('data/alice.txt').read()
wc_obj = wordcloud.WordCloud()
wc = wc_obj.generate(text)

import matplotlib.pyplot as plt

plt.figure()
plt.imshow(wc)
plt.axis('off')
plt.show()
plt.savefig('wc_alice.png')
```

```python
from PIL import Image
import numpy as np
alice_mask = np.array(Image.open('data/alice_mask.png'))
plt.figure()
plt.imshow(alice_mask)
plt.axis("off")
plt.show()
```

```python
import wordcloud
wc_obj = wordcloud.WordCloud(mask=alice_mask)
wc = wc_obj.generate(text)


plt.figure()
plt.imshow(alice_mask)
plt.imshow(wc)
plt.axis("off")
plt.show()
```
